# Systematic Quantification of Observed Data for COGS 13: Assignment 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
observational_data = pd.read_excel('Sitting Checksheet.xlsx')
pd.set_option('display.max_columns', None)
observational_data
```

```python
alone_one_to_two_chairs = len(observational_data[observational_data["Sitting Alone"] == 'x'])
alone_three_plus_chairs = len(observational_data[observational_data["Unnamed: 3"] == 'x'])
```

```python
no_people_alone_seating_label = ('1 to 2 chairs', '3 to 4 or more chairs')
index1 = np.arange(len(no_people_alone_seating_label))
no_people_alone_seating = [alone_one_to_two_chairs, alone_three_plus_chairs]
plt.bar(index1, no_people_alone_seating, align='center', alpha=0.5)
plt.xlabel('Number of Chairs at Table')
plt.ylabel('Number of People Sitting Alone')
plt.xticks(index1, no_people_alone_seating_label)
plt.title('Individuals Sitting Alone at Table vs. Number of Chairs')
plt.show()
```

```python
three_to_four_alone = observational_data[observational_data["Unnamed: 3"] == 'x']
```

```python
alone_adjacent_loners = len(three_to_four_alone[three_to_four_alone["Unnamed: 9"] == 'x'])
alone_adjacent_another_group = len(three_to_four_alone[three_to_four_alone["Unnamed: 10"] == 'x'])
alone_adjacent_no_one = len(three_to_four_alone[three_to_four_alone["Unnamed: 11"] == 'x'])
```

```python
adjacent_alone_label = ['Other people/\n another person \n sitting alone', 'People sitting with \n one another \n (i.e., in a group \n of 2+ people)', 'No one']
index2 = np.arange(len(adjacent_alone_label))
adjacent_alone = [alone_adjacent_loners, alone_adjacent_another_group, alone_adjacent_no_one]
plt.barh(index2, adjacent_alone, align='center', alpha=0.5)
plt.yticks(index2, adjacent_alone_label)
plt.xlabel('Number of People Sitting at Tables \n With 3-4 Chairs')
plt.ylabel('Whom Individual is Adjacent To')
plt.title('Individuals Sitting Alone at Table With 3-4 Chairs vs. Whom They Are Adjacent To')
plt.savefig('adjacent_alone.png')
plt.show()
```

```python
alone_headphones = len(observational_data[observational_data["Unnamed: 8"] == 'x'])
group_headphones = len(observational_data[observational_data["Unnamed: 19"] == 'x'])
```

```python
headphones_label = ('Sitting Alone', 'Sitting in Group')
index3 = np.arange(len(headphones_label))
no_people_headphones = [alone_headphones, group_headphones]
plt.bar(index3, no_people_headphones, align='center', alpha=0.5)
plt.xlabel("Individual's Sitting Arrangement")
plt.ylabel('Number of Individuals Wearing Headphones')
plt.xticks(index3, headphones_label)
plt.title('Individuals Wearing Headphones vs. Sitting Arrangement')
plt.show()
```

```python
individual_tech_contact = len(observational_data[observational_data["Unnamed: 5"] == 'x'])
individual_food_contact = len(observational_data[observational_data["Unnamed: 6"] == 'x'])
individual_elsewhere_contact = len(observational_data[observational_data["Unnamed: 7"] == 'x'])
```

```python
individual_eye_contact_label = ['Technological Device', 'Food', 'Other']
index4 = np.arange(len(individual_eye_contact_label))
individual_eye_contact_object = [individual_tech_contact, individual_food_contact, individual_elsewhere_contact]
plt.bar(index4, individual_eye_contact_object, align='center', alpha=0.5)
plt.xlabel("What Individual Was Making Eye Contact With")
plt.ylabel('Number of Individuals')
plt.xticks(index4, individual_eye_contact_label)
plt.title('What Individuals Sitting Alone Made Eye Contact With')
plt.show()
```

```python
people_contact = observational_data[observational_data["Unnamed: 15"] == 'x']
tech_contact = observational_data[observational_data["Unnamed: 16"] == 'x']
food_contact = observational_data[observational_data["Unnamed: 17"] == 'x']
elsewhere_contact = observational_data[observational_data["Unnamed: 18"] == 'x']

people_contact_freq = len(people_contact)
tech_contact_freq = len(tech_contact)
food_contact_freq = len(food_contact)
elsewhere_contact_freq = len(elsewhere_contact)
```

```python
eye_contact_label = ['Other People \n in Group', 'Technological \n Device', 'Food', 'Other']
index5 = np.arange(len(eye_contact_label))
eye_contact_object = [people_contact_freq, tech_contact_freq, food_contact_freq, elsewhere_contact_freq]
plt.bar(index5, eye_contact_object, align='center', alpha=0.5)
plt.xlabel("What Individual Was Making Eye Contact With")
plt.ylabel('Number of Individuals')
plt.xticks(index5, eye_contact_label)
plt.title('What Individuals in Groups Made Eye Contact With')
plt.show()
```

```python
people_talk = len(people_contact[people_contact["Unnamed: 24"] == 'x'])
tech_talk = len(tech_contact[tech_contact["Unnamed: 24"] == 'x'])
food_talk = len(food_contact[food_contact["Unnamed: 24"] == 'x'])
elsewhere_talk = len(elsewhere_contact[elsewhere_contact["Unnamed: 24"] == 'x'])
```

```python
talking_rates = [people_talk, tech_talk, food_talk, elsewhere_talk]

index6 = np.arange(len(talking_rates)) 
width = 0.35       
plt.bar(index6, eye_contact_object, width, align='center', alpha=0.5, label='Object Eye Contact \n Was Made With')
plt.bar(index6 + width, talking_rates, width, align='center', alpha=0.5, label='Talking')

plt.ylabel('Number of Individuals')
plt.xlabel('Talking Rate of Individuals in Groups and What Individuals Made Eye Contact With')
plt.title('What Individuals Made Eye Contact With and Corresponding Talking Rates')

plt.xticks(index6 + width / 2, ('Other People \n in Group', 'Technological \n Device', 'Food', 'Other'))
plt.legend(loc='best')
plt.show()
```

```python
body_angled = len(observational_data[observational_data["Unnamed: 20"] == 'x'])
sitting_in_group_1_2_chairs = len(observational_data[observational_data["Sitting with Others"] == 'x'])
sitting_in_group_3_4_chairs = len(observational_data[observational_data["Unnamed: 13"] == 'x'])
sitting_in_group_classroom = len(observational_data[observational_data["Unnamed: 14"] == 'x'])
sitting_in_group_total = sitting_in_group_1_2_chairs + sitting_in_group_3_4_chairs + sitting_in_group_classroom
body_not_angled = sitting_in_group_total - body_angled
```

```python
angled_label = ['Body Angled \n Towards Group', 'Body Not Angled \n Towards Group']
index6 = np.arange(len(angled_label))
angled = [body_angled, body_not_angled]
plt.bar(index6, angled, align='center', alpha=0.5)
plt.xlabel("Body Angled/Not Angled Towards Group")
plt.ylabel('Number of Individuals')
plt.xticks(index6, angled_label)
plt.title('Individuals in Groups and Body Alignment Towards Group Members')
plt.show()
```
